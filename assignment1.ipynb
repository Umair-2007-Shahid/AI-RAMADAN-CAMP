```python
print("----- MARKSHEET -----")

m1 = float(input("Enter marks of Subject 1: "))
m2 = float(input("Enter marks of Subject 2: "))
m3 = float(input("Enter marks of Subject 3: "))
m4 = float(input("Enter marks of Subject 4: "))
m5 = float(input("Enter marks of Subject 5: "))

total_marks_obtained = m1 + m2 + m3 + m4 + m5
percent_value = (total_marks_obtained / 500) * 100

print("Total Marks =", total_marks_obtained)
print("Percentage =", percent_value, "%")

if percent_value >= 90:
    grade_value = "A+"
elif percent_value >= 80:
    grade_value = "A"
elif percent_value >= 70:
    grade_value = "B"
elif percent_value >= 60:
    grade_value = "C"
elif percent_value >= 50:
    grade_value = "D"
else:
    grade_value = "Fail"

print("Grade =", grade_value)
```

```python
print("---- SIMPLE CALCULATOR ----")

first_value = float(input("Enter first number: "))
operation_symbol = input("Enter operator (+, -, *, /): ")
second_value = float(input("Enter second number: "))

if operation_symbol == "+":
    answer = first_value + second_value
elif operation_symbol == "-":
    answer = first_value - second_value
elif operation_symbol == "*":
    answer = first_value * second_value
elif operation_symbol == "/":
    if second_value != 0:
        answer = first_value / second_value
    else:
        answer = "Error! Division by zero."
else:
    answer = "Invalid operator!"

print("Result =", answer)
```
